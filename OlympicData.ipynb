# Olimpiyatlara katılan türk atletlerin yaş dağılımı

```python
turkish_athletes =olympic_data[olympic_data['NOC']=='TUR']
turkish_athletes_age=turkish_athletes['Age'].dropna()
turkish_athletes_age
```

```python
plt.figure(figsize=(10, 6)) # boyutlandırma
plt.hist(turkish_athletes_age, edgecolor='black') # edgecolor --> çubuk kenarlığı
plt.title('Olimpiyatlara Katılan Türk Atletlerin Yaş Dağılımı') 
plt.xlabel('Yaş') 
plt.ylabel('Atlet Sayısı') 
plt.grid(axis='y', alpha=0.75) # ızgara çizgileri, alpha --> saydamlık
plt.show() # Grafiği göster
```

```python
age_counts = turkish_athletes_age.value_counts().sort_index()
```

```python
plt.figure(figsize=(14, 7)) 
age_counts.plot(kind='bar', edgecolor='black') 
plt.title('Olimpiyatlara Katılan Türk Atletlerin Yaşa Göre Dağılımı') 
plt.xlabel('Yaş') 
plt.ylabel('Atlet Sayısı') 
plt.xticks(rotation=45, ha='right') # x ekseni etiketlerini eğik yaz (eğer çok fazlaysa)
plt.grid(axis='y', alpha=0.75) 
plt.tight_layout() # grafiğin sıkışmamasını sağlar
plt.show() # grafiği göster
```
