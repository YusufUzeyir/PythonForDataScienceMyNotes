```python
# merge
```

```python
df_merge1=pd.read_csv('7-merge_data1.csv')
```

```python
df_merge1
```

```python
df_merge2=pd.read_csv('7-merge_data2.csv')
```

```python
df_merge2
```

```python
# merge - outer join
```

```python
df_merge_outer=pd.merge(df_merge1,df_merge2, on='Employee_ID',how='outer')
```

```python
df_merge_outer
```

```python
# merge - left join
```

```python
df_merge_left=pd.merge(df_merge1,df_merge2,on='Employee_ID',how='left')
```

```python
df_merge_left
```

```python
# merge - right join
```

```python
df_merge_right=pd.merge(df_merge1,df_merge2,on='Employee_ID',how='right')
```

```python
df_merge_right
```

```python
# merge - inner join
```

```python
df_merge_inner=pd.merge(df_merge1,df_merge2, on='Employee_ID',how='inner')
```

```python
df_merge_inner
```
