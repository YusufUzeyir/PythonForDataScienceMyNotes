```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.head()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex", data=data) # hue --> renk ile kategorik ayrım yapmayı sağlar
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
data.columns
```

```python
data["Medal"].unique()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Medal", data=data) # hue --> renk ile kategorik ayrım yapmayı sağlar
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex",style="Medal", data=data) # hue --> renk ile kategorik ayrım yapmayı sağlar
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x='Height',y='Weight', hue="Sex",style="Medal",size="Age", data=data) # hue --> renk ile kategorik ayrım yapmayı sağlar
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x='Height',y='Weight', hue="Sex", data=data)
plt.xlabel('Height of Atheletes')
plt.ylabel('Weight of Atheletes')
plt.title('Atheletes Weight vs Height')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', hue="Sex",data=data)
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x='Height', hue="Sex",data=data, kind="kde")
plt.ylabel('Frequency')
plt.title('Atheletes Height Histogram')
plt.show()
```

```python
sns.barplot(x='Medal', y='Height', hue="Sex", data=data)
plt.title('Medals by Height')
plt.show()
```

```python
sns.catplot(x='Medal', y='Height', hue='Sex', col='Season', data=data, kind='bar')
```
