```python
def salary_category(salary):
    if salary<50000:
        return "Low"
    elif 50000<= salary <80000:
        return "Medium"
    else:
        return "High"
```

```python
df['Salary_Category']=df['Salary'].apply(salary_category)
```

```python
# deneyimi 10 yÄ±ldan fazla olanlarÄ±n performans skorlarÄ±nÄ± 1 artÄ±rÄ±p yeni sÃ¼tunda listele
```

```python
def calculate(experience):
    if experience >10:
        return 1
    else:
        return 0
```

```python
df['New_Performance_Score']=df['Experience'].apply(calculate)
```

```python
df
```

```python
df['New_Performance_Score_Orjinal']=df['Performance_Score']+df['New_Performance_Score']
```

```python
df
```

```python
# ikinci cozum
```

```python
def adjust_performance(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    return row["Performance_Score"]
```

```python
df["Adjusted_Performance"] = df.apply(adjust_performance, axis=1)
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x: x.replace("_", " "))
```
