```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,85]
```

```python
weight_list = [40,60,80,85,86,87,70,90,95,100]
```

```python
plt.plot(age_list, weight_list,"y")
# plt.savefig("test.png") kaydetmek i√ßin
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
# numpy
```

```python
np_age_list=np.array(age_list)
```

```python
np_age_list
```

```python
np_weight_list=np.array(weight_list)
```

```python
np_weight_list
```

```python
plt.plot(np_age_list,np_weight_list,'g')
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Aeg vs Weight")
plt.show()
```

```python
numpy_array1=np.linspace(0,10,20)
```

```python
numpy_array1
```

```python
numpy_array2=numpy_array1**3
```

```python
numpy_array2
```

```python
plt.plot(numpy_array1, numpy_array2)
plt.show()
```

```python
plt.plot(numpy_array1, numpy_array2,"b*-") # grafik √ßizgi g√∂sterimleri -->    *- , -- , * , -
plt.show()
```

```python
plt.plot(numpy_array2, numpy_array1,"b*-") 
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_array1, numpy_array2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_array2, numpy_array1,"g--")
plt.show()
```

```python
my_figure=plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.4,0.4])
figure_axes.plot(numpy_array1,numpy_array2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig=plt.figure(dpi=100)

new_axes=new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_array1,numpy_array1**2,label="numpy array **2")
new_axes.plot(numpy_array1,numpy_array1**3,label="numpy array **3")
new_axes.legend(loc=9)
plt.show()
```
